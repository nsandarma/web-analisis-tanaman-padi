```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
import plotly.express as px
import plotly
%matplotlib inline
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns',None)
from babel import numbers
# pd.options.display.float_format = '{:,.2f}'.format
```

```python
df = pd.read_csv('dataset/data_lengkap.csv')
print("Berikut Tampilan Data")
total_produksi = sum(df.Produksi)
df.sample(10)
```

```python
total_produksi = sum(df.Produksi)
total_produksi = numbers.format_decimal(total_produksi)
```

```python
print(f"Dataset memiliki {df.shape[0]} baris dan {df.shape[1]} Kolom")
```

```python
df.info()
```

```python
table_check = pd.DataFrame(
    {
        'Data Kosong': df.isnull().sum(),
        'Data Duplikat':df.duplicated().sum(),
        'Data NaNN':df.isna().sum(),
        'Tipe Data':df.dtypes
    }
)
table_check
```

```python
df.corr()
```

```python
#cek korelasi kuat antara Produksi dan Luas Panen
fig = px.scatter(df, x="Luas Panen", y="Produksi",template='plotly_dark', color="Provinsi")
fig.show()
plotly.offline.plot(fig, filename='grafik/scatter_plot.html')
```

```python
total_produksi_prov = df.groupby(['Provinsi'])['Produksi'].sum().sort_values(ascending = False).reset_index()

fig = px.bar(df, y=total_produksi_prov['Produksi'], x=total_produksi_prov['Provinsi'], text_auto='.2s',template='plotly_dark',
             color = total_produksi_prov['Provinsi'])
fig.update_layout(title='Distribusi Produksi Padi di Pulau Sumatera',
                   xaxis_title='Provinsi',
                   yaxis_title='Jumlah Produksi'
                 )
fig.show()
plotly.offline.plot(fig, filename='grafik/distribusi_produksi.html')
```

```python
timeline_prov = df.groupby(['Provinsi','Tahun'] )['Produksi'].sum().reset_index()

fig = px.line(x = timeline_prov['Tahun'], y = timeline_prov['Produksi'],color =timeline_prov['Provinsi'], template='plotly_dark', markers=True)
fig.update_layout(title='Produksi Padi Dari Tahun 1993-2020',
                   xaxis_title='Tahun',
                   yaxis_title='Jumlah Produksi')
fig.show()
a = fig.to_html()
plotly.offline.plot(fig, filename='grafik/time_series.html')

```

```python
C_mat = df.corr()
fig = plt.figure(figsize = (8,8))

sns.heatmap(C_mat, square = True, annot=True)
plt.show()
plt.savefig('corr.png')
```

```python
fig = px.pie(df,values='Produksi',names='Provinsi',title=f'Persentase Distribusi Produksi ({min(df.Tahun)} - {max(df.Tahun)})',template='plotly_dark')
fig.update_traces(hoverinfo='label+percent',textposition='inside', textinfo='percent+label',textfont_size=11)
fig.show()

plotly.offline.plot(fig, filename='grafik/persentasi_distribusi.html')
```

```python
fig = px.imshow(df.corr(),text_auto=True,title='Kolerasi Antar Variabel',template='plotly_dark')
fig.show()
```
